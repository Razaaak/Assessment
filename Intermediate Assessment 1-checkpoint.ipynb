```python
from sklearn.datasets import fetch_california_housing 
import pandas as pd 
housing = fetch_california_housing(as_frame=True) 
df = housing.frame 
print(df.head())
```

```python
#2. Display first 5 rows
print (df.head())
```

```python
#3. check data types and show coloumn name
print(df.dtypes)
```

```python
#4. find missing values
print(df.isnull().sum())
```

```python
#5.statistical summary
print(df.describe())
```

```python
#6. Duplicated values
print(df.duplicated().sum())
```

```python
# To remove
df=df.drop_duplicates()

```

```python
#7. Handle missing values
df.fillna(df.mean(),inplace = True)
```

```python
#8. Add new column price per room
print(df.columns)
```

```python
df['PricePerRoom'] = df['MedHouseVal'] / df['AveRooms']
```

```python
#9. add new column High population area
df['HighPopulationArea'] = df['Population'].apply(lambda x: 1 if x > 500 else 0)
```

```python
df.columns
```

```python
#10.Bin the median_income into 5 equal-sized bins and label them as Very Low, Low, Medium, 
High, Very High.
df['IncomeCategory'] = pd.cut(df['MedInc'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
```

```python
df.columns
```

```python
#11. drop
df = df.drop(['AveRooms', 'Population'], axis=1)
```

```python
df.columns
```

```python
#12.Histogram
import matplotlib.pyplot as plt

plt.hist(df['MedHouseVal'], color='blue', edgecolor='black')
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#13. Sactter plot
plt.scatter(df['Longitude'], df['Latitude'], c=df['MedHouseVal'])
plt.colorbar(label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
#14. Box plot
import seaborn as sns

sns.boxplot(x='IncomeCategory', y='MedHouseVal', data=df)
plt.title('Median House Value by Income Category')
plt.xlabel('Income Category')
plt.ylabel('Median House Value')
plt.show()
```

```python
#17 Heat map
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numerical_df = df.select_dtypes(include=['number']) 
correlation_matrix = numerical_df.corr()  


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()



```

```python
#16.Bar plot
import matplotlib.pyplot as plt

# Group by HighPopulationArea and calculate the mean
avg_values = df.groupby('HighPopulationArea')['MedHouseVal'].mean()

avg_values.plot(kind='bar', color=['orange', 'green'])
plt.title('Average Median House Value: High vs Low Population Areas')
plt.xlabel('High Population Area')
plt.ylabel('Average Median House Value')
plt.show()
```

```python
import seaborn as sns

# Create the pairplot
sns.pairplot(df, vars=['MedInc', 'HouseAge', 'MedHouseVal'])
plt.suptitle('Pairplot of Selected Features', y=1.02)  
plt.show()

```
